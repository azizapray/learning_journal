```python
%matplotlib inline
import random

# Membuat list untuk kepala dan ekor
heads_tails = [0,0]

# loop sebanyak 50000 percobaan melempar
trials = 50000
trial = 0
while trial < trials:
    trial = trial + 1
    # Nomor random 0 dan 1
    # 0 = kepala | 1 = ekor
    toss = random.randint(0,1)
    # Menambahkan komponen dalam list berdasarkan hasil random toss
    heads_tails[toss] = heads_tails[toss] + 1

print (heads_tails)

# Pie chart hasil coin toss
from matplotlib import pyplot as plt
plt.figure(figsize=(5,5))
plt.pie(heads_tails, labels=['heads', 'tails'])
plt.legend()
plt.show()
```
