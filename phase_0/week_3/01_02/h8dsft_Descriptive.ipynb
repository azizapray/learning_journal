# Covid Statistics

```python
import pandas as pd
```

## Data Loading

```python
covid = pd.read_csv('covid_19_indonesia_time_series_all.csv')
covid
```

## Preprocessing Data

```python
filter_cov = covid.loc[:, :'Total Active Cases']
filter_cov
```

```python
filter_cov1 = filter_cov[
    (filter_cov['Location'] == 'Jawa Timur')
]

filter_cov1
```

```python
filter_cov1.info()
```

```python
filter_cov1['Date'] = pd.to_datetime(filter_cov1['Date'])
filter_cov1.info()
```

## Processing Data

### Measures of Central Tendency

```python

print("Berikut ini adalah statistik\nPertambahan Kasus Baru di Provinsi Jawa Timur")
print("Sejak", filter_cov1['Date'].min(), "hingga", filter_cov1['Date'].max())
# mean
print("\nMean:", int(filter_cov1['New Cases'].mean()))
# median
print("Median:", int(filter_cov1['New Cases'].median()))
# mode
print("Mode:", filter_cov1['New Cases'].mode()[0])
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

n_cases = filter_cov1['New Cases']
n_cases.plot.hist(title='Salary Distribution', color='lightblue', bins=50)  
plt.axvline(n_cases.mean(), color='magenta', linestyle='dashed', linewidth=2)
plt.axvline(n_cases.median(), color='green', linestyle='dashed', linewidth=2)
plt.show()
```

#### Measures of Variance

```python
# Range
numcols = ['Total Cases', 'Total Deaths', 'Total Recovered']
for col in numcols:
    print(filter_cov1[col].name + ' range: ' + str(filter_cov1[col].max() - filter_cov1[col].min()))
```

```python
# Quartiles
print("Quartiles:\n",n_cases.quantile([1/4, 2/4, 3/4]))
```

```python
# Outliers with box-whisker chart
filter_cov1['New Cases'].plot(kind='box', title='Jawa Timur Covid New Cases Distribution', figsize=(10,8))
plt.show()
```

```python
# Variance
print("\nVariance:", n_cases.var())

# Std. Deviation
print("Std. Deviation:", n_cases.std())
```
