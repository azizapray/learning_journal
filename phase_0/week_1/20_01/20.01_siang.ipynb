## Querring your dataset

```python
df.head()
```

```python
df[
    (df['year_id'] > 1950) &
    (df['fran_id'] == 'Knicks') &
    (df['pts'] > 150)
]
```

```python
df[
    (df['year_id'] == 1953) &
    (df['team_id'] == 'BLB') &
    (df['pts'] > 100) &
    (df['opp_pts'] > 100)
]
```

```python
df.query("team_id == 'BLB' & pts > 100 & opp_pts > 100 & year_id == 1953")
```

## Grouping and Aggregating Data

```python
df.groupby('fran_id')['pts'].sum()
```

```python
df[
    (df['year_id'] > 2010) &
    (df['fran_id'] == 'Knicks')
].groupby('year_id')['pts'].sum()
```

```python
df[
    (df['year_id'] > 2010) &
    (df['fran_id'] == 'Knicks')
].groupby(['year_id', 'game_location'])['pts'].sum()
```
