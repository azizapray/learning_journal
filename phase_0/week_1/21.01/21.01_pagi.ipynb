## Function, Looping, dan If

hitung **luas** jika dual_sim = 1, jika tidak maka hitung **kurang**:

- luas = tinggi(px_height)* lebar(px_width)
- kurang = tinggi -lebar
- buat kolom baru yang berisi hasil dari perhitungan luas dan kurang

```python
def luas(tinggi,lebar):
    hasil = tinggi * lebar
    return hasil

def kurang(tinggi,lebar):
    hasil_k = tinggi - lebar
    return hasil_k
```

```python
df_new = df.copy()
```

```python
lk = []     # untuk masukan hasil kedalam list
g = 0      # indeks baris

for i in df_new['dual_sim']:
    if i == 1:
        lk.append(luas(df_new['px_height'][g], df_new['px_width'][g]))
    else:
        lk.append(luas(df_new['px_height'][g], df_new['px_width'][g]))
    print(df_new['px_height'][g])
    g += 1

df_new['l/k'] = lk      # bikin kolom 'l/k' dengan nilai list 'lk'
```
